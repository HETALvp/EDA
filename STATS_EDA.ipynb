```python
def get_outliers(series):
  outliers = []
  Q1 = series.quantile(.25)
  Q3 = series.quantile(.75)
  IQR = Q3 - Q1
  print(IQR)
  LL = Q1 - 1.5 * IQR
  print(LL)
  UL = Q3 + 1.5 * IQR
  print(UL)

  for value in series:
    if value < LL or value > UL:
      outliers.append(value)
  print(len(outliers))
  return outliers


print(get_outliers(df['Revenue']))
```
