```python
import pandas as pd
import matplotlib.pyplot as plty
import seaborn as sns
```

```python
#load data
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
# @title Year vs Laps

from matplotlib import pyplot as plt
data.plot(kind='scatter', x='Year', y='Laps', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Year vs Laps

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Year']
  ys = series['Laps']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = data.sort_values('Year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Team_nationality')):
  _plot_series(series, series_name, i)
  fig.legend(title='Team_nationality', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('Laps')
```

IOR

- IQR = Q3 - Q1
- LL = Q1 - 1.5IQR
- UL = Q3 + 1.5IQR

```python
Q1 = temperatures.quantile(.25)
print(Q1)
Q3 = temperatures.quantile(.75)
print(Q3)
IQR = Q3 - Q1
print(IQR)
```

```python
def get_outliers(series):
  outliers = []
  Q1 = series.quantile(.25)
  Q3 = series.quantile(.75)
  IQR = Q3 - Q1
  LL = Q1 - 1.5 * IQR
  print(LL)
  UL = Q3 + 1.5 * IQR
  print(UL)

  for value in series:
    if value < LL or value > UL:
      outliers.append(value)
  return outliers


print(get_outliers(temperatures))
```
