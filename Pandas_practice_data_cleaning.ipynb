```python
df = bios.copy()
```

```python
df.head()
```

####1.Refactor name column

```python
df['name'] = df['Used name'].str.replace('', ' ')
df.insert(4, 'name', df.pop('name')) # removed `name` column form last position to 4th place
df.head()
```

```python
#practice
# splited_measures = df.at[1,'Measurements'].split('/')
# splited_measures
```

```python
# df.drop(columns=['height_cm', 'weight_kg'], inplace=True)
df[['height_cm', 'weight_kg']] = df['Measurements'].str.split('/', expand = True)
# df['weight_kg'].fillna(df['height_cm'], inplace=True) #good idea to replace weight_kg nan values with the height itself

# df['height_cm'] = pd.Series([x if ('cm' in str(x)) else np.nan for x in df['height_cm']]) #bad idea, weight_kg gets replaced by none, how will i get the original then?

df['weight_kg'] = pd.to_numeric(df['weight_kg'].str.strip(' kg'), errors='coerce')
df['height_cm'] = pd.to_numeric(df['height_cm'].str.strip(' cm'), errors='coerce')

df.head()
```

```python
df[~df['Measurements'].str.contains('cm', na=False) & df['Measurements'].notnull()]
```

```python
df.sort_values(by='height_cm', ascending=False)
```

####3.Extract born and died date

```python
df['born_date'] = df['Born'].str.extract(r'(\d+ \w+ \d{4}|\d{4})')
df['born_date'] = pd.to_datetime(df['born_date'], format='mixed', errors='coerce')
df.head()
```

```python
df['died_date'] = df['Died'].str.extract(r'(\d+ \w+ \d{4})| \d{4}')
df['died_date'] = pd.to_datetime(df['died_date'], format='mixed', errors='coerce')
df.head()
```
