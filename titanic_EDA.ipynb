```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
```

```python
#import train data
titanic_data = pd.read_csv('train.csv')
titanic_data.head()
```

```python
titanic_data.info()
```

```python
titanic_data.isnull().sum()
```

```python
#looking at null values, will drop column cabin and fillNa with mean in AGE column
titanic_data.drop(columns='Cabin', inplace=True)
titanic_data['Age'].fillna(titanic_data['Age'].mean, inplace=True)
titanic_data.isnull().sum()
```

```python
#encode gender (Label Encoder method)
# le = LabelEncoder()
# titanic_data['Sex'] = le.fit_transform(titanic_data['Sex'])
# titanic_data.head()

#get_dummies method
titanic_data = pd.get_dummies(columns=['Sex'], data=titanic_data, drop_first=True)
titanic_data.head()
```
